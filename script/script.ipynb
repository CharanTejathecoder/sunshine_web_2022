```python
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx')
```

```python
dept_dict = {
    "ai": "Artificial Intelligence",
    "bm": "Biomedical Engineering",
    "bo": "",
    "ce": "Civil Engineering",
    "cs": "Computer Science",
    "cy": "Chemical Engineering",
    "ee": "Electrical Engineering",
    "ma": "Mathematics",
    "md": "Design",
    "me": "Mechanical Engineering",
    "ms": "Materials Science",
    "sm": "Smart Mobility"
}
```

```python
with open('output.txt', 'w') as f:
    for dept in dept_dict.keys():
        f.write(f"""<div class="row justify-content-md-center">
            <div class=" col-md-12  col-sm-12 student-header">
               <h2 class="branch  sm-center"><b>{dept_dict[dept]}</b></h2>
            </div>
         </div>\n""")
        f.write("""<div class="row justify-content-md-center"><center>""")
        df_inner = df[list(map(lambda x: x.startswith(dept), df["Email Address"]))]
        if df_inner.shape[0] == 1:
            flag = 4
        elif df_inner.shape[0] == 3:
            flag = 2
        f.write(f"<div class=\"col-md-{flag} col-sm-12\"></div>")
        for index, row in df_inner.iterrows():
            print(row['Roll no.'])
            f.write(f"""<div class="col-md-4 col-sm-12">
                  <div class="student">
                     <div class="imageHolder">
                        <img src="images/team2021/pg/IMG_20210114_091042 - Jagadeesh Mahadevan.jpg" class="student-img img-responsive">
                     </div>
                     <br>{row['Name']}<br><br> {row['Email Address']}
                  </div>
               </div>""")
        f.write("""</center>
         </div>""")
        
```
